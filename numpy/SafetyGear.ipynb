```python
import matplotlib.pyplot as plt
```

```python
transdata=data.transpose()
```

```python
transdata
```

```python
transdata[0]
```

```python
plt.plot(transdata[0],transdata[1])
plt.ylabel('Acc(m/s2)')
plt.xlabel('Time(s)')
plt.title("Acc-Time")
plt.grid(True)
plt.show()
```

```python
import matplotlib as mpl

%matplotlib inline
plt.style.use('bmh')
line_1,=plt.plot(transdata[0],transdata[1],label='Acc Line')
line_2,=plt.plot(transdata[0],transdata[2],label='Vel Line')
line_3,=plt.plot(transdata[0],transdata[3],label='Dis Line',color='red')
plt.legend(handles=[line_1,line_2,line_3])
plt.xlabel('Time(s)')
plt.grid(True)
plt.savefig('12.png',dpi=150)
plt.show()
```

```python
plt.style.use('bmh')
plt.figure(1)
plt.grid(True)
plt.subplot(311)
plt.plot(transdata[0],transdata[1],label='Acc Line')
plt.subplot(312)
plt.plot(transdata[0],transdata[2],label='Vel Line')
plt.subplot(313)
plt.plot(transdata[0],transdata[3],label='Dis Line')
plt.savefig('11.png',dpi=150)
plt.show()
```
