```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as pl
import matplotlib
import math
```

```python
N=500
fs=5
n = [2*math.pi*fs*t/N for t in range(N)] 
axis_x=np.linspace(0,1,num=N)
```

```python
x=[math.sin(i) for i in n]
pl.subplot(221)
pl.plot(axis_x,x)
pl.title('5Hz sin signal')
pl.axis('tight')
pl.show()
```

```python
xx=[]
x1=[math.sin(i*10) for i in n]
for i in range(len(x)):
    xx.append(x[i]+x1[i])
    
```

```python
pl.subplot(222)
pl.plot(axis_x,xx)
pl.title('5Hz-50Hz sin signal')
pl.axis('tight')
pl.show()
```

```python
b,a=signal.butter(3,0.08,'low')
sf=signal.filtfilt(b,a,xx)
```

```python
pl.subplot(223)
pl.plot(axis_x,sf)
pl.title('low filter')
pl.axis('tight')
pl.show()
```

```python
b,a=signal.butter(3,0.10,'high')
sf=signal.filtfilt(b,a,xx)
```

```python
pl.subplot(224)
pl.plot(axis_x,sf)
pl.title('high filter')
pl.axis('tight')
pl.show()
```
